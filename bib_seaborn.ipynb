# <span style='color:red'>Seaborn : Data Visualization sous Python</span>

#### Definition:
Seaborn est une librairie de visualisation de données Python basée sur Matplotlib. Elle fournit une interface plus conviviale pour la création de certaines visualisations de données, notamment les diagrammes de dispersion et les histogrammes.

```python
# importation des bibliotheques
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# importer le jeu de donnees
from sklearn.datasets import load_iris
x = load_iris().data
y = load_iris().target
data = load_iris()
iris = pd.DataFrame(x, columns=load_iris().feature_names)
iris.head()
```

```python
# afficher les nuages de point
sns.scatterplot(iris["sepal length (cm)"], iris['sepal width (cm)'])
```

```python
# les diagrammes en bar
sns.barplot(y, iris['sepal length (cm)'])
```
